```python
from matplotlib import pyplot as plt
```

```python
labels=['Python','C++','Ruby','Java','R']
sizes=[250,130,200,210,160]
colors=['Yellow','Yellowgreen','Pink','Blue','Red']
exp=[0,0,0,0,.2]
plt.pie(sizes,explode=exp,labels=labels,colors=colors,startangle=90,shadow=True,autopct='%.2f%%')#Pie circle banata hai
plt.legend()
plt.show()
```

```python
col=['red','green','blue']
siz=[54,22,24]
val=['veg','non','not']
exp=[0,0,0]
plt.pie(siz,labels=val,colors=col,explode=exp,startangle=90)
mcircle=plt.Circle((0,0),.4,color='white')
p=plt.gcf()
p.gca().add_artist(mcircle)
plt.show()
```

```python
import numpy as np
labels=['Python','C++','Ruby','Java','R']
sizes=np.array([250,130,200,210,100])
colors=['Yellow','Yellowgreen','Pink','Blue','Red']
exp=[0,0,0,0,.1]
def calc(val):
    a=np.round(val/100*sizes.sum())
    return a
plt.pie(sizes,explode=exp,labels=labels,colors=colors,startangle=90,shadow=True,autopct=calc)#Pie circle banata hai
plt.legend()
plt.show()
```

```python
x=np.arange(0,3*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
plt.plot(x,y_cos,'r')
plt.plot(x,y_sin,'g')
plt.show()
```
