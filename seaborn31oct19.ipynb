```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
def sinplot(f=1):
    x=np.linspace(0, 20, 100)
    for i in range(1, 7):
        plt.plot(x,np.sin(x+i*.5))
sn.set()
sn.set_style('ticks') #for background dark,darkgrid,whitegrid,ticks
sinplot()
```

```python
def sinplot(f=1):
    x=np.linspace(0, 8, 100)
    for i in range(1,7):
        plt.plot(x,np.sin(x+1*.5))
sn.set()
sn.set_style('dark')
sinplot()
```
